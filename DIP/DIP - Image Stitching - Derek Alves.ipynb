```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
files = ['ic1.jpg', 'ic2.jpg', 'ic3.jpg', 'ic4.jpg']

images = []

for f in files:
  img = cv.imread(f)
  images.append(img)

fig = plt.figure(figsize=(40, 10))

i = 1
for img in images:
  plt.subplot(1, len(images), i)
  plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
  plt.title(files[i-1])
  i = i + 1
```

```python
Stitcher = cv.Stitcher.create()
_, stitched_img = Stitcher.stitch(images)

fig = plt.figure(figsize=(40, 10))
plt.imshow(cv.cvtColor(stitched_img, cv.COLOR_BGR2RGB))
```

```python
cv.imwrite('stit_ic.jpg', stitched_img)
```
