# Questão 4
A imagem ```babbon_lena.png``` contém uma combinação das imagens ```baboon.png``` e ```lena.png``` utilizando a máscara ```mask.png```. Desenvolva o código que efetua essa combinação.

````python
img_1 = cv2.imread('img/baboon.png')
img_2 = cv2.imread('img/lena.png')
img_3 = cv2.imread('img/baboon_lena.png')

img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
img_3 = cv2.cvtColor(img_3, cv2.COLOR_BGR2RGB)

result = np.zeros((1,1))# TODO

# Altere as próximas linhas de código
# TODO

mask = cv2.imread('img/mask.png')
plt.imshow(mask)


def mask_bwa(f, g):
  return cv2.bitwise_and(f, g)

def mask_bwo(f, g):
  return cv2.bitwise_or(f, g)




img_31 = mask_bwa(img_1, mask)
plt.imshow(img_31)

mask_r = cv2.rotate(mask, 2)
````

````python
img_32 = mask_bwa(img_2, mask_r)
plt.imshow(img_32)
````

````python
img_3 = mask_bwo(img_31, img_32)
````
