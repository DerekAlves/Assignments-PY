```python
# Q43

def cov2corr(cov):
  v = np.sqrt(np.diag(cov))
  outer_v = np.outer(v, v)
  corr = cov / outer_v
  corr[cov == 0] = 0
  return corr

arr1 = np.array([3, 1, 2])
print(arr1, '\n\n')
arr2 = np.array([5, -8, 9])
print(arr2, '\n\n')

print(cov2corr(np.cov(arr1, arr2)))
```

```python
# Q101

def descArr(arr):
  mean = np.mean(arr)
  median = np.median(arr)

  min = np.amin(arr)
  max = np.amax(arr)
  range = np.ptp(arr)
  variance = np.var(arr)
  sd = np.std(arr)

  print("Descrição:")
  print("Array =", arr)
  print("Medidas de tendência central")
  print("Média =", mean)
  print("Mediana =", median)
  print("Medidas de dispersão")
  print("Mínimo =", min)
  print("Máximo =", max)
  print("Alcance =", range)
  print("Variância =", variance)
  print("Desvio padrão =", sd)
  return

arr = np.arange(10)
descArr(arr)
```
